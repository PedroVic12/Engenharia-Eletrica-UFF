## Exemplo de numeros aleatorios

```python
import random

def numero_inteiro():
    print("Gerando cinco numeros aleatorios entre 1 e 50:")
    
    array_inteiro = []
    for i in range(1,51):
        numero = random.randint(1,51)

        if numero not in array_inteiro:
            array_inteiro.append(numero)

    array_inteiro.sort()

    print("Tamanho da Lista = ",len(array_inteiro))
    return array_inteiro

numero_inteiro()
```

# Código Principal

```python
def linearMap(array,limite_a, limite_b):
    n = len(array)
    numero_max = max(array)
    numero_min = min(array)

    amplitude = numero_max - numero_min

    diferença_limite = limite_b - limite_a

    coeficiente = diferença_limite/amplitude

    resultado = []

    for i in range(0,n):
        resultado.append(limite_a + (coeficiente*i)) 

    print(n)
    return resultado


lista_original = numero_inteiro()

valores = linearMap(lista_original, 6,8)
valores
```

```python
import matplotlib.pyplot as grafico

grafico.ylabel("Lista Original")
grafico.xlabel("Valores")
grafico.plot(lista_original,valores)
```
