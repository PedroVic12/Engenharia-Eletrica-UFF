```python
import numpy as np
import matplotlib.pyplot as plt

# Funes fornecidas
funcoes = [
    ('f(x) = 1/x', lambda x: 1/x),
    ('f(x) = 1/(x^2 - 1)', lambda x: 1/(x**2 - 1)),
    ('f(x) = (x^3 + 1) / x^3', lambda x: (x**3 + 1) / x**3),
    ('f(x) = (3x^2 + 1) / (2x^2 + 1)', lambda x: (3*x**2 + 1) / (2*x**2 + 1)),
    ('f(x) = x^2 / ((x - 1)(x - 2))', lambda x: x**2 / ((x - 1)*(x - 2))),
    ('f(x) = sqrt(3) / (x^2 + 1) / (x - 1)', lambda x: np.sqrt(3) / ((x**2 + 1) * (x - 1)))
]

for nome, funcao in funcoes:
    # Intervalo de x para plotagem
    x = np.linspace(-10, 10, 400)
    # Valores de y para a funo
    y = funcao(x)
    
    # Plotando o grfico da funo
    plt.plot(x, y, label=nome)
    
    # Encontrando assntotas horizontais
    assintotas_horizontais = [0]
    
    # Encontrando assntotas verticais
    assintotas_verticais = []
    for raiz in np.roots(np.poly1d([1, -3, 0, 1])):
        if np.isreal(raiz):
            assintotas_verticais.append(raiz)
    
    # Plotando as assntotas verticais
    for assintota in assintotas_verticais:
        plt.axvline(x=assintota, color='r', linestyle='--', label='Assntota vertical')
        
    # Adicionando legendas
    plt.title('Grfico de ' + nome)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

```
