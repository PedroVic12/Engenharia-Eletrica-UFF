```python
m1 = np.array([[1, 2], [7, 8]])

m2 = np.array([[4, 5],  [10, 11]])

m3 = np.matrix([[1,2],[3,1]])


Mprod = m1* m2
Mdot = np.dot(m1,m2)

print(f"{m1} x {m2}")

print("Produto de matrizes = \n", Mprod)
print("Produto escalar = \n", Mdot)

```

```python
m1_t = np.transpose(m1)

m1_inversa = np.linalg.inv(m1)

print("Matriz transposta = \n", m1_t)
print("Matriz inversa = \n", m1_inversa)

print("Matriz multiplicada = \n", m1_t.dot(m1_inversa))





# m1_t = np.transpose(m1)
```
