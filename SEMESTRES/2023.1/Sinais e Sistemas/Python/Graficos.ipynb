```python
import numpy as np
import matplotlib.pyplot as plt

# Criando um sinal de entrada
x = np.array([0, 1, 2, 3, 4])

# Criando uma resposta ao impulso
h = np.array([1, 2, 3])

# Calculando a saída do sistema (convolução discreta)
y = np.convolve(x, h)

# Plotando os sinais e a resposta ao impulso
plt.stem(x, use_line_collection=True)
plt.title('Sinal de entrada x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()

plt.stem(h, use_line_collection=True)
plt.title('Resposta ao impulso h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.show()

plt.stem(y, use_line_collection=True)
plt.title('Saída do sistema y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()
```
