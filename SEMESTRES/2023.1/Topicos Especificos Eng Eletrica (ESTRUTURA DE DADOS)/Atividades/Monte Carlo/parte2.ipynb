```python
import random
import matplotlib.pyplot as plt

def power(v):
    return 0.05 * v**3

def average_power():
    num_samples = 1000000  # Número de amostras para a simulação
    sum_power = 0
    power_samples = []

    for _ in range(num_samples):
        v = random.uniform(7, 14)  # Amostragem aleatória da velocidade do vento
        p = power(v)
        sum_power += p
        power_samples.append(p)

    avg_power = sum_power / num_samples
    return avg_power, power_samples

avg_power, power_samples = average_power()

# Plotando o histograma das amostras de potência
plt.hist(power_samples, bins=100, edgecolor='black')
plt.xlabel('Potência (kW)')
plt.ylabel('Frequência')
plt.title('Distribuição da Potência da Usina Eólica')
plt.grid(True)

# Adicionando uma linha vertical para representar a potência média
plt.axvline(x=avg_power, color='red', linestyle='--', linewidth=2, label='Potência Média')
plt.legend()

# Mostrando o gráfico
plt.show()

print("Potência média da usina:", avg_power)
```
