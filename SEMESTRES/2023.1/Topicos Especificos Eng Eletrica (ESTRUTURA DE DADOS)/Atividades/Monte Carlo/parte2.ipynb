## Potencia MÃ©dia de uma turbina

```python
import random
import matplotlib.pyplot as plt

def power(v):
    return 0.05 * v**3

def average_power():
    num_samples = 1000000  # NÃºmero de amostras para a simulaÃ§Ã£o
    sum_power = 0
    power_samples = []

    for i in range(num_samples):
        v = random.uniform(7, 14)  # Amostragem aleatÃ³ria da velocidade do vento
        p = power(v)
        sum_power += p
        power_samples.append(p)

    avg_power = sum_power / num_samples
    return avg_power, power_samples

avg_power, power_samples = average_power()

# Plotando o histograma das amostras de potÃªncia
plt.hist(power_samples, bins=100, edgecolor='black')
plt.xlabel('PotÃªncia (kW)')
plt.ylabel('FrequÃªncia')
plt.title('DistribuiÃ§Ã£o da PotÃªncia da Usina EÃ³lica')
plt.grid(True)

# Adicionando uma linha vertical para representar a potÃªncia mÃ©dia
plt.axvline(x=avg_power, color='red', linestyle='--', linewidth=2, label='PotÃªncia MÃ©dia')
plt.legend()

# Mostrando o grÃ¡fico
plt.show()

print("PotÃªncia mÃ©dia da usina:", avg_power)
```
