## Rotacional em Python

```python
class Calculo2:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def soma(self):
        return self.num1 + self.num2

    def subtracao(self):
        return self.num1 - self.num2

    def multiplicacao(self):
        return self.num1 * self.num2

    def divisao(self):
        return self.num1 / self.num2

    def potencia(self):
        return self.num1 ** self.num2

    def raiz(self):
        return self.num1 ** (1/self.num2)

    def calcula(self):
        print(f'Soma: {self.soma()}')
        print(f'Subtração: {self.subtracao()}')
        print(f'Multiplicação: {self.multiplicacao()}')
        print(f'Divisão: {self.divisao()}')
        print(f'Potência: {self.potencia()}')
        print(f'Raiz: {self.raiz()}')

    def cout(self,valor, n = None):
        if n is None:
            n = self.num1
        print(f'\n{valor} = {n}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

NUM_SETAS = 20

x = np.linspace(0,4,NUM_SETAS)
y = np.arange(0,6,5)

calculadora = Calculo2(x,y)

print("x = ", x)
print("y = ", y)
```

```python

x1,y1 = np.meshgrid(x,y)
V = -x1 * (x1 -4)


calculadora.cout("x1",x1)
calculadora.cout("Y1",y1)
calculadora.cout("Veolcidade",V)


Vx = np.gradient(V)[0]
v_x = V[0,:]
calculadora.cout("Gradiente VX",v_x)
v_y =V
v_x = 0*v_y

plt.quiver(x1,y1,v_x,v_y, scale = 1, scale_units = 'xy', angles = "xy")
plt.ylim(0,10)
plt.grid()
plt.show()
```

```python

dv_dy, dv_dx = np.gradient(v_y,y,x)
calculadora.cout(" F'(x)",dv_dy)
calculadora.cout(" F'(y)",dv_dx)

rotZ = dv_dx - dv_dy
calculadora.cout("rotacional",rotZ)

fig = plt.figure(figsize=(8,5), dpi= 150)
ax1 = plt.axes(projection='3d')
ax1.plot_surface(x1,y1,V, cmap='viridis')

plt.title('Campo Vetorial')
plt.show()
```

```python
fig = plt.figure(figsize=(8,5), dpi= 150)
ax1 = plt.axes(projection='3d')
ax1.quiver(x1,y1,0*x1,v_x,v_y,0*v_x, color = 'green', label = "Velocidade = $v(x,y)$")
ax1.scatter(x1,y1,0*x1, color = 'red')
ax1.set_zlim(-5,5)

ax1.quiver(x1,y1,0*x1,rotZ*0,rotZ*0,rotZ, label = "Rotacional")
ax1.legend()
```
