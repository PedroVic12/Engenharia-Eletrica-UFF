## Tipos de Regime

```python
import numpy as np
import matplotlib.pyplot as plt

class GeradorGrafico:
    def __init__(self):
        self.fig, self.ax = plt.subplots(1, 3, figsize=(15, 5))

    def plot_perfil_velocidade(self, x, perfil, label, titulo, ax_index):
        self.ax[ax_index].plot(x, perfil, label=label)
        self.ax[ax_index].set_xlabel('Posição')
        self.ax[ax_index].set_ylabel('Perfil de Velocidade')
        self.ax[ax_index].set_title(titulo)
        self.ax[ax_index].legend()

    def exibir_graficos(self):
        plt.show()

def calcular_perfil_velocidade(x, reynolds):
    if reynolds < 2000:
        perfil = 6 * x * (1 - x)  # Regime laminar
    elif reynolds < 4000:
        perfil = 2 * x * (1 - x)  # Regime de transição
    else:
        perfil = 0.5 * x * (1 - x)  # Regime turbulento
    return perfil

# Valores de Reynolds para cada regime
reynolds_laminar = 100
reynolds_transicao = 2000
reynolds_turbulento = 5000

# Defina uma série de pontos para o perfil de velocidade
x = np.linspace(0, 1, 100)

# Calcule os perfis de velocidade para cada regime
perfil_laminar = calcular_perfil_velocidade(x, reynolds_laminar)
perfil_transicao = calcular_perfil_velocidade(x, reynolds_transicao)
perfil_turbulento = calcular_perfil_velocidade(x, reynolds_turbulento)

# Crie uma instância da classe GeradorGrafico
gerador = GeradorGrafico()

# Plote o perfil de velocidade para o regime laminar
gerador.plot_perfil_velocidade(x, perfil_laminar, 'Laminar', f'Regime Laminar (Re = {reynolds_laminar})', 0)

# Plote o perfil de velocidade para o regime de transição
gerador.plot_perfil_velocidade(x, perfil_transicao, 'Transição', f'Regime de Transição (Re = {reynolds_transicao})', 1)

# Plote o perfil de velocidade para o regime turbulento
gerador.plot_perfil_velocidade(x, perfil_turbulento, 'Turbulento', f'Regime Turbulento (Re = {reynolds_turbulento})', 2)

# Exiba os gráficos
gerador.exibir_graficos()
```
