```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros dados
rho = 1000  # kg/m³
mu = 10  # [Pa.s] = [kg/(m.s)]
L = 1  # m (como exemplo)

# Velocidades do escoamento
U = np.linspace(0, 10, 100)  # m/s

# Cálculo do número de Reynolds
Re = (rho * U * L) / mu

# Criação do gráfico
plt.figure(figsize=(10, 6))
plt.plot(U, Re, label='Número de Reynolds vs. Velocidade')
plt.axhspan(0, 2000, color='green', alpha=0.3, label='Regime Laminar')
plt.axhspan(2000, 4000, color='yellow', alpha=0.3, label='Regime de Transição')
plt.axhspan(4000, max(Re), color='red', alpha=0.3, label='Regime Turbulento')

# Adicionando títulos e rótulos
plt.title('Número de Reynolds vs. Velocidade do Escoamento')
plt.xlabel('Velocidade do Escoamento (m/s)')
plt.ylabel('Número de Reynolds')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros dados
rho = 1000  # kg/m³
mu = 10  # [Pa.s] = [kg/(m.s)]
L = 1  # m

def calcular_Re(U, rho, mu, L):
    """Calcula o Número de Reynolds para uma dada velocidade U."""
    return (rho * U * L) / mu

def plotar_Re(U, Re, titulo):
    """Cria um gráfico do Número de Reynolds em função da velocidade do escoamento."""
    plt.plot(U, Re, label='Número de Reynolds vs. Velocidade')
    plt.title(titulo)
    plt.xlabel('Velocidade do Escoamento (m/s)')
    plt.ylabel('Número de Reynolds')
    plt.grid(True)
    plt.legend()
    plt.show()

# Faixas de Velocidade para cada regime
U_laminar = np.linspace(0, 1, 100)  # m/s
U_transicao = np.linspace(1, 2.5, 100)  # m/s
U_turbulento = np.linspace(2.5, 10, 100)  # m/s

# Cálculo do Número de Reynolds para cada regime
Re_laminar = calcular_Re(U_laminar, rho, mu, L)
Re_transicao = calcular_Re(U_transicao, rho, mu, L)
Re_turbulento = calcular_Re(U_turbulento, rho, mu, L)

# Criação dos Gráficos
plotar_Re(U_laminar, Re_laminar, 'Regime Laminar')
plotar_Re(U_transicao, Re_transicao, 'Regime de Transição')
plotar_Re(U_turbulento, Re_turbulento, 'Regime Turbulento')
```
