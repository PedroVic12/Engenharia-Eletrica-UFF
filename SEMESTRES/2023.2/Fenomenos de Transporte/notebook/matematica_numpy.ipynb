```python
import numpy as np

# Velocidades medidas em um experimento (m/s)
velocidades = np.array([1.2, 1.3, 1.1, 1.25, 1.35])

# Cálculo da média e desvio padrão
media = np.mean(velocidades)
desvio_padrao = np.std(velocidades)

print(f"Média: {media:.2f} m/s, Desvio Padrão: {desvio_padrao:.2f} m/s")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
velocidade = np.array([0.5, 1.0, 1.5, 2.0, 2.5])
pressao = np.array([1.2, 2.8, 4.5, 6.0, 7.8])

# Ajuste de uma curva polinomial de segundo grau
coeficientes = np.polyfit(velocidade, pressao, 2)
polinomio = np.poly1d(coeficientes)

# Plotagem
velocidade_fina = np.linspace(0, 3, 100)
plt.scatter(velocidade, pressao, label='Dados Experimentais')
plt.plot(velocidade_fina, polinomio(velocidade_fina), label='Ajuste Polinomial', linestyle='--')
plt.xlabel('Velocidade (m/s)')
plt.ylabel('Pressão (Pa)')
plt.legend()
plt.show()
```

```python
from scipy.integrate import odeint

# Definição da EDO
def modelo(y, t, k):
    dydt = -k * y
    return dydt

# Condição inicial
y0 = 5

# Tempo onde a solução será calculada
t = np.linspace(0, 10, 100)

# Constante k
k = 0.5

# Resolvendo a EDO
solucao = odeint(modelo, y0, t, args=(k,))

# Plotagem
plt.plot(t, solucao, label=f'k={k}')
plt.xlabel('Tempo')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
